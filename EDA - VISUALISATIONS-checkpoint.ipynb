```python
CB = pd.read_csv("Centre_Backs.csv")
```

```python
for i in CB.columns:
    print(i)
```

```python
sns.set_theme(style = "darkgrid")
ax = sns.countplot(x = CB.preferred_foot)
for label in ax.containers:
    ax.bar_label(label)
plt.title("Number of Right and Left footed Centre Backs")
plt.show()
```

```python
CB_h_p = pd.DataFrame(CB[["height_cm", "physic"]])
```

```python
CB_h_p.groupby("height_cm").mean().plot.line()
plt.title("Height vs Physic among Centre Backs")
plt.show()
```

```python
CB_p_d = pd.DataFrame(CB[["physic", "defending"]])
```

```python
CB_p_d.groupby("physic").mean().plot.line()
plt.title("Physic vs Defending among Centre Backs")
plt.show()
```

```python
CB_pj_aha = pd.DataFrame(CB[["power_jumping", "attacking_heading_accuracy"]])
```

```python
CB_pj_aha.groupby("power_jumping").mean().plot.line()
plt.title("Jumping vs Attacking headers")
```
